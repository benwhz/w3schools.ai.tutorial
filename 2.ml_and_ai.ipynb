## 2.1 什么是人工智能

人工智能表明机器可以模仿人类：

- 说话
- 思考
- 学习
- 规划
- 理解

人工智能也称为机器智能和计算机智能。

- Arthur Samuel 1959:
> “机器学习是计算机科学的一个分支，它使计算机无需编程就能学习” 
>
> <span style="font-size:0.7em;">*Arthur Samuel, IBM Journal of Research and Development, Vol. 3, 1959.*</span>

- Wikipedia 2022：

人工智能是机器所展现的智能。与人类和动物所展现的自然智能不同，后者涉及意识和情感。

- Investopedia 2022：

人工智能是指在机器中模拟人类智能，这些机器被编程为像人类一样思考并模仿他们的行为。

- IBM 2022：

人工智能利用计算机和机器来模仿人类思维的解决问题和决策能力。

- Britannica 2022：

人工智能是数字计算机或计算机控制的机器人执行通常与智能生物相关的任务的能力，......例如推理、发现意义、概括或从过去的经验中学习的能力。

因此人工智能是一门涵盖从狭义人工智能到强人工智能的多个数据科学领域的科学学科，包括机器学习、深度学习、大数据和数据挖掘。

## 2.2 弱人工智能

狭义人工智能仅限于狭义（特定）领域，就像我们今天身边的大多数人工智能一样：

- 电子邮件垃圾邮件过滤器
- 文本转语音
- 语音识别
- 自动驾驶汽车
- 电子支付
- 谷歌地图
- 文本自动更正
- 自动翻译
- 聊天机器人
- 社交媒体
- 人脸检测
- 视觉感知
- 搜索算法
- 机器人
- 自动投资
- NLP - 自然语言处理
- 飞行无人机
- IBM 的 Dr. Watson
- 苹果的 Siri
- 微软的 Cortana
- 亚马逊的 Alexa
- Netflix 的推荐

狭义人工智能也称为弱人工智能。

弱人工智能：旨在模拟人类智能。

强人工智能：旨在复制人类智能。

<a name='2'></a>
## 2.3 强人工智能

强人工智能是模仿人类智能的人工智能类型。

强人工智能表示思考、计划、学习和交流的能力。

强人工智能是人工智能的理论下一个层次：真正的智能。

强人工智能朝着具有自我意识、意识和客观思想的机器发展。

> 我们不需要判断机器是否能够“思考”。我们只需要判断机器是否能够像人类一样聪明地行动。
>
> <span style="font-size:0.7em;">*Alan Turing*</span>
