## 10.1 关系

机器学习系统使用输入到输出之间的关系来产生预测的。

在代数中，关系通常写为：$y = ax + b$：

- $y$ 是我们要预测的标签
- $a$ 是直线的斜率
- $x$ 是输入值
- $b$ 是截距

对于机器学习来说，关系写为：$y = b + wx$：

- $y$ 是我们要预测的标签
- $w$ 是权重（斜率）
- $x$ 是特征（输入值）
- $b$ 是截距

<a name='2'></a>
## 10.2 标签

在机器学习术语中，标签就是我们想要预测的答案。它就像线性图中的 $y$ 值：

|代数|机器学习|
|:----:|:-------:|
|**y** = ax + b|**y** = b + wx|

<a name='3'></a>
## 10.3 特征

在机器学习术语中，特征就是输入。它们就像线性图中的 $x$ 值：

|代数|机器学习|
|:----:|:-------:|
|y = a**x** + b|y = b + w**x**|

有时可能有多个具有不同权重（$w_1, w_2, ..., w_n$）的特征（输入值）：

$y = b + w_1x_1 + w_2x_2 + w_3x_3 + ... + w_nx_n$

<a name='4'></a>
## 10.4 模型

模型定义了标签 ($y$) 和特征 ($x$) 之间的关系。

模型的生命周期分为三个阶段：

- 数据收集
- 训练
- 推理

模型通常由以下内容组成：

- 神经网络结构定义
    - 网络层次
    - 网络结构
    - 输入和输出
- 模型基础参数
- 模型权重参数

<a name='5'></a>
## 10.5 训练

训练的目标是创建一个可以正确推理出答案的模型。例如，房屋的预期价格是多少？

<a name='6'></a>
## 10.6 推理

推理是指使用训练后的模型通过实时数据推断（预测）值。就像将模型投入生产一样。

<a name='7'></a>
## 10.7 阶段

机器学习分两个主要阶段：

1. 训练：
输入数据用于计算模型的参数。

2. 推理：
“训练过的”模型接受任何输入推导出正确答案的过程。

<a name='8'></a>
## 10.8 监督学习

监督式机器学习使用一组输入变量来预测输出变量的值。

监督式学习使用**标记数据**（具有已知答案的数据）来训练算法以：

- 对数据进行分类
- 预测结果

监督式学习可以根据已知的垃圾邮件示例对数据进行**分类**，例如“电子邮件中的垃圾邮件是什么”。

监督式学习可以根据您播放过的视频来**预测**结果，例如预测您喜欢哪种类型的视频。

<a name='9'></a>
## 10.9 无监督学习

无监督机器学习使用来自任何未标记数据集的模式，试图理解数据中的模式（或分组）。

无监督学习用于预测未定义的关系，如数据中的有意义的模式。

它是关于创建可以自我改进的计算机算法。

预计机器学习将转向无监督学习，以允许程序员在不创建模型的情况下解决问题。

<a name='10'></a>
## 10.10 强化学习

强化学习基于非监督学习，但会从用户那里获得反馈，无论决策是好是坏。反馈有助于改进模型。

<a name='11'></a>
## 10.11 自监督学习

自监督学习与无监督学习类似，因为它处理的数据没有人工添加的标签。

不同之处在于，无监督学习使用聚类、分组和降维，而自监督学习针对回归和分类任务得出自己的结论。
