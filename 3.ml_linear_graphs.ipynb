机器学习经常使用线性图来显示变量之间的关系，如用直线图显示线性函数的值：$y = ax + b$

重要关键字：

- **线性**（直线）
- **斜率**（角度）
- **截距**（起始值）

<a name='1'></a>
## 3.1 线性

线性说明是直线关系，因此线性图是一条直线。

线性图由两个轴组成：x 轴（水平）和 y 轴（垂直）。

```python
# importing the required libraries
import matplotlib.pyplot as plt
import numpy as np
 
# define data values
x = np.arange(0, 10, 1)
y = x
 
plt.plot(x, y)
plt.title('y = x')
plt.show()
```

<a name='2'></a>
## 3.2 斜率

斜率是直线倾斜的角度，斜率是线性图形中的 $a$ 值，如：$y = ax$

例如：以下示例中，斜率 = 1.2：

```python
y = x*1.2
 
plt.plot(x, y)
plt.title('y = 1.2x')
plt.show()
```

<a name='3'></a>
## 3.3 截距
截距是直线图在y轴上的起始值，截距是线性图形中的 $b$ 值，如：$y = ax + b$

例： 在以下示例中，斜率 = 1.2，截距 = 7：

```python
y = x*1.2 + 7
 
plt.plot(x, y)
plt.title('y = 1.2x + 7')
plt.show()
```
