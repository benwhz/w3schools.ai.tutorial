## 5.1 开创者

弗兰克·罗森布拉特 (1928 – 1971) 是一位在人工智能领域颇有名气的美国心理学家。1957 年，他开始做了一件大事。他在康奈尔航空实验室的 IBM 704 计算机上“发明”了一个感知器程序。科学家发现，脑细胞（神经元）通过电信号接收来自我们感官的输入。神经元则使用电信号来存储信息，并根据先前的输入做出决策。弗兰克认为感知器可以模拟大脑原理，具有学习和决策的能力。

## 5.2 感知器

最初的感知器被设计用来接收多个二进制输入，并产生一个二进制输出（0 或 1），其理念是使用不同的权重来表示每个输入的重要性，并且在做出是或否（真或假）（0 或 1）之类的决定之前，这些值的总和应该大于阈值。

## 5.3 感知器实例

想象出一个感知器（在你的大脑中），利用该感知器来决定你是否应该去听音乐会。

艺术家好吗？天气好吗？这些事实应该具有什么权重？

|判断准则|输入|权重|
|--|--|--|
|艺术家优秀吗？|x1=0或1|w1=0.7|
|天气情况好吗？|x2=0或1|w2=0.6|
|朋友一起去吗？|x3=0或1|w3=0.5|
|提供美食吗？|x4=0或1|w4=0.3|
|提供酒精饮料吗？|x5=0或1|w5=0.4|

<a name='4'></a>
## 5.4 感知器算法

弗兰克·罗森布拉特提出了以下算法：
1. 设置一个阈值，如：$Threshold = 1.5$
2. 将所有输入与其权重相乘：
- $x1 * w1 = 1 * 0.7 = 0.7$
- $x2 * w2 = 0 * 0.6 = 0$
- $x3 * w3 = 1 * 0.5 = 0.5$
- $x4 * w4 = 0 * 0.3 = 0$
- $x5 * w5 = 1 * 0.4 = 0.4$
3. 将所有结果相加：
- $0.7 + 0 + 0.5 + 0 + 0.4 = 1.6 (The Weighted Sum)$
4. 激活输出：
- 如果总和 > 1.5 (“是的，我会去听音乐会”)，则返回 true

**注意：**
<div class=notes>
如果对于你来说，天气权重是 0.6，那么对于其他人来说，情况可能有所不同。权重越高意味着天气对他们来说越重要。

如果对于你来说，阈值是 1.5，那么对于其他人来说，情况可能有所不同。阈值越低意味着他们更想去听音乐会。
</div>

下面是感知器的计算示例：

```python
import numpy as np

threshold = 1.5
x = np.array([1,0,1,0,1])
w = np.array([0.7,0.6,0.5,0.3,0.4])

y = w*x
if y.sum() > threshold:
    print("Yes, I will go to the Concert.")
else:
    print("Sorry, I'm not going.")   
```

## 5.5 人工智能中的感知器

**感知器**是一种人工神经元，它受到**生物神经元**功能的启发。

它在人工智能中起着至关重要的作用，它是神经网络中的重要组成部分。

要了解其背后的理论，我们可以分解其组成部分：

- 感知器输入（节点）
- 节点值（1、0、1、0、1）
- 节点权重（0.7、0.6、0.5、0.3、0.4）
- 激活函数
- 阈值
- 求和（总和 > 阈值）

### 5.5.1 感知器输入
感知器接收一个或多个输入，感知器输入称为节点，节点既有值又有权重。
### 5.5.2 节点值
输入节点的二进制值为 1 或 0，这可以解释为真或假/是或否。
在上面的示例中，节点值为：<span class=important>1、0、1、0、1</span>。
### 5.5.3 节点权重
权重是分配给每个输入的值，权重显示每个节点的强度，值越高意味着输入对输出的影响越大。

在上面的例子中，节点权重为：<span class=important>0.7、0.6、0.5、0.3、0.4</span>。
### 5.5.4 求和
感知器计算其输入的加权和，它将每个输入乘以其相应的权重，然后将结果相加。

在上面的例子中，总和为：<span class=important>1x0.7 + 0x0.6 + 1x0.5 + 0x0.3 + 1x0.4 = 1.6</span>。
### 5.5.5 激活函数
求和后，感知器将激活函数应用于总和，目的是将非线性引入输出。它根据聚合输入确定感知器是否应触发。

在上面的例子中，激活函数很简单：<span class=important>（总和 > 1.5）</span>。
### 5.5.6 阈值
激活函数通常伴随着阈值,如果激活函数的结果超过阈值，感知器就会触发（输出 1），否则它保持不活动状态（输出 0）。

在上面的例子中，阈值是：<span class=important>1.5</span>。
### 5.5.7 输出
感知器的最终输出是激活函数的结果,它表示感知器基于输入和权重的决策或预测。

激活函数将加权和映射到二进制值,二进制 1 或 0 可以解释为真或假/是或否。

在上面的例子中，输出为 <span class=important>1</span>，因为：（总和 > 1.5）或（1.6 > 1.5）。

## 5.6 感知器的学习

感知器可以通过称为训练的过程从示例中学习。

在训练期间，感知器根据观察到的误差调整其权重。这通常使用学习算法（例如感知器学习规则或反向传播算法）来完成。

学习过程向感知器提供标记的示例，其中期望输出是已知的。感知器将其输出与期望输出进行比较，并相应地调整其权重，旨在最大限度地减少预测输出和期望输出之间的误差。

学习过程允许感知器学习权重，使其能够对新的未知输入做出准确的预测。

<div class=notes>
显然，决策不能仅由一个神经元做出。

其他神经元必须提供更多输入：

- 艺术家好吗

- 天气好吗
...

**多层感知器**可用于更复杂的决策。
</div>

值得注意的是，虽然感知器在人工神经网络的发展中发挥了重要作用，但它们仅限于学习线性可分离模式。

然而，通过将多个感知器分层堆叠在一起并结合非线性激活函数，神经网络可以克服这一限制并学习更复杂的模式。

## 5.7 神经网络

感知器定义了进入神经网络的第一步：

<img alt="Neural Networks" style="width:100%;max-width:400px" src="https://www.w3schools.com/ai/img_neural_networks.jpg">

感知器通常用作更复杂的神经网络的构建块，例如多层感知器 (MLP) 或深度神经网络 (DNN)。

通过分层组合多个感知器并将它们连接到网络结构中，这些模型可以学习和表示数据中的复杂模式和关系，从而实现图像识别、自然语言处理和决策等任务。
