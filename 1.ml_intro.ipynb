## 1.2 神经网络

神经网络是：
- 一种编程技术
- 机器学习中使用的一种方法
- 一种能从错误中学习的软件框架

神经网络是受人脑的工作方式的启发：神经元相互发送信息。当人脑试图解决问题时（神经元反复发送信息），它会加强导致成功的连接并削弱导致失败的连接。

## 1.4 多层神经网络

神经网络是具有多层结构的感知器集合。

<img alt="Neural Networks" style="width:100%;max-width:300px" src="https://www.w3schools.com/ai/img_nn_single_600.jpg">

最简单的神经网络由以下部分组成：

- 输入层（黄色）
- 隐藏层（蓝色）
- 输出层（红色）

在**神经网络**模型中，输入数据（黄色）在产生最终输出（红色）之前，会通过隐藏层（蓝色）进行处理。

**第一层：**
黄色感知器根据输入做出简单决策。每个决策都会发送到下一层中的感知器。

**第二层：**
蓝色感知器通过权衡第一层的结果来做出决策，这一层比第一层在更抽象的层面上做出更复杂的决策。

<a name='5'></a>
## 1.5 深度神经网络

深度神经网络由多个隐藏的神经网络层组成，可对大量数据执行复杂的操作。
每个连续层都使用前一层作为输入。

例如，光学阅读使用低层来识别边缘，使用高层来识别字母。

<img alt="Neural Networks" style="width:100%;max-width:300px" src="https://www.w3schools.com/ai/img_neural_networks.jpg">

在**深度神经网络**模型中，输入数据（黄色）在隐藏层（蓝色）上进行处理，并在更多隐藏层（绿色）上进行修改，以产生最终输出（红色）。

**第一层：**
黄色感知器根据输入做出简单决策。每个决策都会发送到下一层的感知器。

**第二层：**
蓝色感知器通过权衡第一层的结果来做出决策。这一层在比第一层更抽象的层面上做出更复杂的决策。

**第三层：**
绿色感知器做出更复杂的决策。

## 1.6 深度学习

深度学习是机器学习的一个子集, 也是近年来人工智能蓬勃发展的原因。深度学习是一种高级机器学习，可以处理图像识别等复杂任务。

下面是机器学习和深度学习的比较：

|机器学习|深度学习|
|--|--|
|人工智能的子集|机器学习的子集|
|通常用小型数据集进行训练|一般使用超大型数据集进行训练|
|通常需要人类指导|除了人类指导外，也可以自己学习|
|生成比较简单的算法|生成的算法比较复杂|
