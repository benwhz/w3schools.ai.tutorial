## 4.1 数据收集

收集数据是任何机器学习项目中最重要的部分，最常收集的数据就是数字和测量值，数据通常存储在表示值之间关系的数组中。

如下表包含房价与面积之间的关系数据：

|价格 |	7	|8	|8|	9|	9	|9|	10|	11|	14|	14|	15|
|-|-|-|-|-|-|-|-|-|-|-|-|
|面积|	50|	60|	70|	80|	90|	100|	110|	120|	130|	140|	150|

<a name='2'></a>
## 4.2 散点图

散点图在二维坐标系上通过散布的点表示两个值之间的关系，例如在下面的示例中通过散点图表示房屋价格和面积之间的对应关系。

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([50,60,70,80,90,100,110,120,130,140,150])
y = np.array([7,8,8,9,9,9,10,11,14,14,15])

plt.title('scatter (Price / Size)')
plt.scatter(x, y)
plt.show()
```

<a name='3'></a>
## 4.3 折线图

同样也可以用折线图表示上述数据之间的关系。

```python
plt.title('plotline (Price / Size)')
plt.plot(x, y)
plt.show()
```

<a name='4'></a>
## 4.4 散点图用途

散点图非常适合：
- 查看“数据总体变化趋势”
- 比较不同的值
- 发现潜在趋势
- 发现数据中的模式
- 发现数据之间的关系
- 发现聚类和相关性
