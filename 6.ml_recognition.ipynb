## 6.2 编程实现

我们可以使用一个简单的 Python 程序来实现上述问题，它将：

1. 创建一个简单的绘图仪
2. 创建 500 个随机 (x, y) 点
3. 显示 (x, y) 点
4. 创建直线函数：func(x)
5. 显示线
6. 计算正确答案
7. 显示正确答案

<a name='21'></a>
### 6.2.1 生成随机数据

创建任意数量的 xy 点。

让 x 值随机（介于 0 和最大值之间）。

让 y 值随机（介于 0 和最大值之间）。

在绘图仪中显示点并且定义线性函数 $y=1.2x+50$ 并以绿线绘制。

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randint(500, size=(2, 500))

def func(x):
    return 1.2*x + 50

plt.title('random data')
plt.scatter(data[0], data[1])

x = np.arange(0, 500, 1)
y = func(x)
 
plt.plot(x, y, color='green')
plt.ylim(0, 500)
```

<a name='22'></a>
### 6.2.2 计算正确值

根据线函数计算正确答案：

$y = x * 1.2 + 50$

如果 y 超过线，则期望答案为 1，如果 y 低于线，则期望答案为 0。

将期望答案存储在数组 (desired[]) 中。

```python
desired = []

for i in range(500):
    if(data[1][i] > func(data[0][i])):
        desired.append(1)
    else:
        desired.append(0)
```

<a name='23'></a>
### 6.2.3 绘制正确值

对于每个点，如果 desire[i] == 1，则显示一个黑点，否则显示一个蓝点。

```python
plt.plot(x, y, color='green')
plt.ylim(0, 500)

for i in range(len(desired)):
    if desired[i] == 1:
        plt.plot(data[0][i],data[1][i],'x', color = "black")
    else:
        plt.plot(data[0][i],data[1][i],'o', color = "blue")
```
